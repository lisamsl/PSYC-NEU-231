## Read in data from eeg study
* This time I've kept in data across a 1s window from many electrodes (20) sampled at 250Hz, so now we have a [960, 20, 250] matrix that contains data from all 960 trials, 20 electrodes, and over a 1s window following stimulus onset
* Also read in a "cond_labels" vector that has 960 0's and 1's to mark trials from the two experimental conditions 

## import stuff...

```python
import numpy as np
from scipy.linalg import eigh
from scipy.linalg import sqrtm
import matplotlib.pyplot as plt
from sklearn.svm import SVC # This is the library for the Support Vector Machine implementation
```

```python
# load the data...
eeg = np.load('/Users/lisasletten/Documents/GitHub/class_demo_eeg.npz')

# get the different arrays like this...
eeg['data']
eeg['sr']
eeg['tx']
eeg['cond_labels']

# shape of data
print('Shape of the big eeg data set: ', eeg['data'].shape)

# and if you want to save some typing, especially because we only have a few variables, you reassign the different arrays like this
data = eeg['data']
sr = eeg['sr']
tx = eeg['tx']
cond_labels = eeg['cond_labels']
```

## Compute the mean across the first dimension (trials) and then plot the mean response across time in each of the 20 electrodes on one axis

* Might need to transpose the data into a Time x Electrode format

```python
#Computing mean across 1st dimension (use axis=0)

dmean =np.mean(data, axis=0)

#Plotting, using transpose of mean

plt.plot(tx, dmean.T)
plt.show()
```

## That gives you an idea about what the pattern of data looks like across electrodes. 
* Next, take the mean across time, so you have a 960 x 20 matrix 
* Use the cond_labels vector to sort the data based on condition (so you have two 480 x 20 matrices)
* Do classification to see if you can decode the experimental condition
* Remember - break up your data into independent training and testing sets
* Set C to a small number if this is taking a long time...

```python
# Taking the mean across time, axis=2
tmean = np.mean(data, axis=2)

#Sorting the data using the condition labels
data0 = tmean[cond_labels==0,:]
data1 = tmean[cond_labels==1,:]

#Splitting data into training data (90%) and test data (10%)
#Define the training data to be 90%
data_trn_p = .9
#Find the number of corresponding points by multiplying percentage with data size 
data_no = int(np.floor(data_trn_p * data1.shape[0]))


#Using vstack to stack vectors vertically (row-wise). Pick out sufficient no of training data
#Separating test data and training data
data_trn = np.vstack((data0[:data_no,:], data1[:data_no,:]))
trn_labels = np.hstack((np.zeros(data_no), np.ones(data_no)))

data_tst = np.vstack((data0[data_no:,:], data1[data_no:,:]))
data_no_t = int(data_tst.shape[0]/2)
tst_labels = np.hstack((np.zeros(data_no_t), np.ones(data_no_t)))
                       
#Using SVC and setting C to a small number                       
model = SVC(kernel='linear', C=2)
model.fit(data_trn, trn_labels)

                       
#We then go on to use the model on the test data
                       
res = model.predict(data_tst) #(the labels of the prediction)
accuracy = np.sum(res==tst_labels) /  data_tst.shape[0]
print('Classification accuracy with a SVM is: ', accuracy)
```

## Now try doing this across timepoints!
* First do a time window based on where you think there is nice separation in the responses across electrodes (look at the data that you plotted above)...maybe try 300-400msec post stim
* Then you can try timepoint x timepoint or a moving window

```python
# Window from 300 ms after stimulus to 400 ms

#We caculate the mean in time over the window
mean_t = np.mean(data[:,:,((tx <= 300)& (tx <= 400))], axis=2)

#Sorting the data using the condition labels
data0 = mean_t[cond_labels==0,:]
data1 = mean_t[cond_labels==1,:]

#Splitting data into training data (90%) and test data (10%)
#Define the training data to be 90%
data_trn_p = .9
#Find the number of corresponding points by multiplying percentage with data size 
data_no = int(np.floor(data_trn_p * data1.shape[0]))

#Using vstack to stack vectors vertically (row-wise). Pick out sufficient no of training data
#Separating test data and training data
data_trn = np.vstack((data0[:data_no,:], data1[:data_no,:]))
trn_labels = np.hstack((np.zeros(data_no), np.ones(data_no)))

data_tst = np.vstack((data0[data_no:,:], data1[data_no:,:]))
data_no_t = int(data_tst.shape[0]/2)
tst_labels = np.hstack((np.zeros(data_no_t), np.ones(data_no_t)))
                       
#Using SVC and setting C to a small number                       
model = SVC(kernel='linear', C=2)
model.fit(data_trn, trn_labels)

#We then go on to use the model on the test data
                       
res = model.predict(data_tst) #(the labels of the prediction)
accuracy = np.sum(res==tst_labels) /  data_tst.shape[0]
print('Classification accuracy with a SVM is: ', accuracy)

```
